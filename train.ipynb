```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df = pd.read_csv('dataset.csv')
df.columns = [i for i in range(df.shape[1])]
df
```

```python
df = df.rename(columns={63: 'Output'})
df
```

```python
X = df.iloc[:, :-1]
print("Features shape =", X.shape)

Y = df.iloc[:, -1]
print("Labels shape =", Y.shape)

X = X.values
Y =Y.values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import LabelBinarizer


# Create an instance of LabelBinarizer
label_binarizer = LabelBinarizer()

# Fit and transform the labels to one-hot encoding
Y_tr = label_binarizer.fit_transform(y_train)
Y_te = label_binarizer.fit_transform(y_test)

print("Original labels:")
print(Y_tr)

print("\nOne-hot encoded labels:")
print(Y_tr)
print(Y_te)
```

```python
Y_tr.shape
```

```python
from keras import layers, models
from keras.callbacks import EarlyStopping


# Define the Conv1D model
model = models.Sequential()
model.add(layers.Conv1D(32, 3, activation='relu', input_shape=(63, 1)))
model.add(layers.MaxPooling1D(2))
model.add(layers.Conv1D(64, 3, activation='relu'))
model.add(layers.MaxPooling1D(2))
model.add(layers.Conv1D(128, 3, activation='relu'))
model.add(layers.MaxPooling1D(2))
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))  

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)


# Train the model
history = model.fit(x_train, Y_tr, epochs=500, batch_size=32, validation_split=0.1,callbacks=[early_stopping])
```

```python
model.evaluate(x_test,Y_te)
```

```python
model.summary()
```

#Enhancing Our model CNN
# increased the number of filters in the Conv1D layers, changed the kernel size, added a dropout layer for regularization, and increased the number of units in the dense layers

```python
from keras import layers, models
from keras.callbacks import EarlyStopping

# Define the Conv1D model with adjusted hyperparameters
model1 = models.Sequential()
model1.add(layers.Conv1D(64, 5, activation='relu', input_shape=(63, 1)))  # Increased filters and kernel size
model1.add(layers.MaxPooling1D(2))
model1.add(layers.Conv1D(128, 5, activation='relu'))  # Increased filters and kernel size
model1.add(layers.MaxPooling1D(2))
model1.add(layers.Conv1D(256, 5, activation='relu'))  # Increased filters and kernel size
model1.add(layers.MaxPooling1D(2))
model1.add(layers.Flatten())
model1.add(layers.Dense(256, activation='relu'))  # Increased units
model1.add(layers.Dropout(0.5))  # Added dropout for regularization
model1.add(layers.Dense(128, activation='relu'))  # Added another dense layer
model1.add(layers.Dense(10, activation='softmax'))  

# Compile the model
model1.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model
history1 = model1.fit(x_train, Y_tr, epochs=500, batch_size=32, validation_split=0.1, callbacks=[early_stopping])
```

```python
model1.summary()
```

```python
model1.evaluate(x_test,Y_te)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping


model_lstm = Sequential()
model_lstm.add(LSTM(64, return_sequences=True, activation='relu', input_shape=(63,1)))
model_lstm.add(LSTM(128, return_sequences=True, activation='relu'))
model_lstm.add(LSTM(64, return_sequences=False, activation='relu'))
model_lstm.add(Dense(64, activation='relu'))
model_lstm.add(Dense(32, activation='relu'))
model_lstm.add(Dense(10, activation='softmax'))

model_lstm.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)


history_lstm = model_lstm.fit(x_train, Y_tr, epochs=500, batch_size=32, validation_split=0.1,callbacks=[early_stopping])
```

```python
model_lstm.summary()
```

##Adding the dopout layer ,increasing the number of units

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping

model_lstm1 = Sequential()
model_lstm1.add(LSTM(128, return_sequences=True, activation='relu', input_shape=(63, 1)))  # Increased LSTM units
model_lstm1.add(Dropout(0.5))  # Added dropout for regularization
model_lstm1.add(LSTM(256, return_sequences=True, activation='relu'))  # Increased LSTM units
model_lstm1.add(Dropout(0.5))  # Added dropout for regularization
model_lstm1.add(LSTM(128, return_sequences=False, activation='relu'))  # Increased LSTM units
model_lstm1.add(Dropout(0.5))  # Added dropout for regularization
model_lstm1.add(Dense(128, activation='relu'))  # Increased dense layer units
model_lstm1.add(Dropout(0.5))  # Added dropout for regularization
model_lstm1.add(Dense(64, activation='relu'))  # Increased dense layer units
model_lstm1.add(Dense(10, activation='softmax'))

model_lstm1.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

history_lstm1 = model_lstm1.fit(x_train, Y_tr, epochs=500, batch_size=32, validation_split=0.1, callbacks=[early_stopping])
```

```python
import matplotlib.pyplot as plt

def plot_loss_curves(history):
  loss = history.history['loss']
  val_loss = history.history['val_loss']
  accuracy = history.history['accuracy']
  val_accuracy = history.history['val_accuracy']
  epochs = range(len(history.history['loss']))

  #plot loss
  plt.figure()
  plt.plot(epochs, loss, label='training_loss')
  plt.plot(epochs, val_loss, label='val_loss')
  plt.title("loss")
  plt.xlabel("epochs")
  plt.legend()

   #plot accuracy
  plt.figure()
  plt.plot(epochs, accuracy, label='training_accuracy')
  plt.plot(epochs, val_accuracy, label='val_accuracy')
  plt.title("accuracy")
  plt.xlabel("epochs")
  plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import itertools

def make_confusion_matrix(y_true, y_pred, classes=None, figsize=(10, 10), text_size=15):
    # Assuming y_true and y_pred are arrays with binary values for each label
    n_classes = y_true.shape[1]

    # create the confusion matrix
    cm = confusion_matrix(np.argmax(y_true, axis=1), np.argmax(y_pred, axis=1))
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]  # normalize our confusion matrix

    # lets prettify it
    fig, ax = plt.subplots(figsize=figsize)
    # Create a matrix plot
    cax = ax.matshow(cm, cmap=plt.cm.Blues)
    fig.colorbar(cax)

    if classes is not None:
        labels = classes
    else:
        labels = np.arange(n_classes)

    # Label the axes
    ax.set(title='Confusion Matrix',
           xlabel='predicted label',
           ylabel='True label',
           xticks=np.arange(n_classes),
           yticks=np.arange(n_classes),
           xticklabels=labels,
           yticklabels=labels)

    # Set x_axis labels to bottom
    ax.xaxis.set_label_position("bottom")
    ax.xaxis.tick_bottom()

    # Adjust label size
    ax.yaxis.label.set_size(text_size)
    ax.xaxis.label.set_size(text_size)
    ax.title.set_size(text_size)

    # set threshold for different colors
    threshold = (cm.max() + cm.min()) / 2.

    # Plot the text on each cell
    for i, j in itertools.product(range(n_classes), range(n_classes)):
        plt.text(j, i, f"{cm[i, j]} ({cm_normalized[i, j] * 100:.1f}%)",
                 horizontalalignment='center',
                 color='white' if cm[i, j] > threshold else 'black', size=text_size)
```
